```python
def retrieve_keys(
    path='/Users/colby/Documents/Lambda/03 Unit 3/twitterapi.keys'
):
    """Retrieves my twitter api keys because .env files won't work"""

    file = open(
        '/Users/colby/Documents/Lambda/03 Unit 3/twitterapi.keys',
        'r'
    )

    data = file.read().split('\n')

    keys = {}

    for x in data:
        y = x.split('=')

        keys[y[0]] = y[1]

    return keys

def twit_connect(keys=retrieve_keys()):
    """Connects to twitter. Returns a tweepy.API object."""

    auth = tweepy.OAuthHandler(keys['KEY'], keys['SECRET'])
    twitter = tweepy.API(auth)

    return twitter
```

```python
twitter = twit_connect()
```

```python
jeff = twitter.get_user('jeffbezos')
nasa = twitter.get_user('nasa')
```

```python
jefftwee = jeff.timeline(count=200, exclude_replies=True, include_rts=False, tweet_mode='Extended')
nasatwee = nasa.timeline(count=200, exclude_replies=True, include_rts=False, tweet_mode='Extended')
```

```python
jeff_tweets_vec = np.array([vec(tweet.text) for tweet in jefftwee])
nasa_tweets_vec = np.array([vec(tweet.text) for tweet in nasatwee])
```

```python
jeff_tweets_vec
```

```python
print(len(nasa_tweets_vec))
print(len(jeff_tweets_vec))
```

```python
print(nasa_tweets_vec.shape)
print(jeff_tweets_vec.shape)
```

```python
vects = np.vstack([nasa_tweets_vec, jeff_tweets_vec])
```

```python
vects.shape
```

```python
labels = np.concatenate([np.zeros(len(nasatwee)), np.ones(len(jefftwee))])
labels
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression().fit(vects, labels)
```

```python
some_text = "Let's go to the moon!"
```

```python
some_text_vec = nlp(some_text).vector
len(some_text_vec)
```

```python
some_text_vec = some_text_vec.reshape(1, -1)
```

```python
log_reg.predict(some_text_vec)
```

```python
def test_tweet(text):
    vectored = nlp(text).vector
    reshaped = vectored.reshape(1, -1)

    return log_reg.predict(reshaped)[0]
```

```python
print(test_tweet('Bruh'))
```
